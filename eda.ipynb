```python
import pyreadr
import pandas as pd
import os

# -----------------------------------------------
# CAMBIA ESTA RUTA A LA DE TU ARCHIVO
rdata_path = "data/listings.RData"

# Carpeta donde se guardarán los CSV (por defecto, la misma carpeta del archivo)
output_dir = os.path.dirname(os.path.abspath(rdata_path))
# -----------------------------------------------

# Leer el archivo
result = pyreadr.read_r(rdata_path)

print("Objetos encontrados en el archivo:")
for nombre, df in result.items():
    print(f"  - {nombre}: {type(df).__name__}")

# Exportar cada DataFrame a CSV
for nombre, df in result.items():
    if not isinstance(df, pd.DataFrame):
        print(f"'{nombre}' no es un DataFrame, se omite.")
        continue

    csv_path = os.path.join(output_dir, f"{nombre}.csv")
    df.to_csv(csv_path, index=False, encoding="utf-8")
    print(f"✓ '{nombre}' guardado en: {csv_path}")
    print(f"  {df.shape[0]} filas x {df.shape[1]} columnas")
    display(df.head())
```
